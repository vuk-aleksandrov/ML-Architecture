```python
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X**4 + X**2 + np.random.randn(100, 1)
```

```python
print(y)
```

```python
# Normal equation

X_b = np.c_[np.ones((100, 1)), X]

#theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# this one is using pseudo-inverse-matrix
theta_best = np.linalg.pinv(X_b).dot(y)

X_new = np.array([[0], [3]])
Xnewb = np.c_[np.ones((2, 1)), X_new]
y_predict = Xnewb.dot(theta_best)

plt.plot(X_new, y_predict, "g-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
# Batch/Full Gradient descent

eta = 0.1 #learning rate
n_iterations = 1000

theta = np.random.randn(2, 1)
m = 100

for iteration in range(n_iterations):
    gradient = 2/m * X_b.T.dot(X_b.dot(theta)-y)
    theta = theta - eta * gradient
    print(X_b.dot(theta).shape)
X_new = np.array([[0], [3]])
Xnewb = np.c_[np.ones((2, 1)), X_new]
y_predict = Xnewb.dot(theta)

plt.plot(X_new, y_predict, "g-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
# Ridge Regression

from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha=1, solver='cholesky')
ridge_reg.fit(X, y)

x=np.array(range(0, 100))

y_predict = ridge_reg.predict(x)

print(y_predict)
```
