```python
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

m = 100
X = 6 * np.random.rand(m, 1) - 3
y = np.sin(X) + 2 + np.random.rand(m, 1)

plt.plot(X, y, 'b.')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree = 2, include_bias=True)
X_poly = poly_features.fit_transform(X)
print(X_poly[0])
```

```python
print(X_poly.shape)
```

```python
# Ridge Regression

alpha = 2
A = np.identity(3)
A[0][0]=0

theta = np.linalg.inv(X_poly.T.dot(X_poly)+ alpha*A).dot(X_poly.T.dot(y))
```

```python
Xnew=np.ndarray(shape=(m, 1), buffer=np.arange(-6, 6, 0.1))
Xnew_poly = poly_features.fit_transform(Xnew)

y_predict = Xnew_poly.dot(theta)

plt.plot(X, y, 'g.')
plt.plot(Xnew, y_predict)
```
