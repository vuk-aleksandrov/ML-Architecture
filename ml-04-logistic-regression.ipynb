```python
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt


from sklearn import datasets

iris = datasets.load_iris()
```

```python
print(list(iris.keys()))
```

```python
X = iris["data"][:, 3:]
y= (iris["target"]==2).astype(int).reshape(150, 1)
```

```python
print(y.shape)
```

```python
def sigmoid(t):
    return 1/(1+np.exp(-t))
```

```python
from sklearn.linear_model import LogisticRegression

eta = 0.1 #learning rate
n_iterations = 1000

theta = np.random.randn(2, 1)
m = 150

X_b = np.c_[np.ones((m, 1)), X]

for iteration in range(n_iterations):
    gradient = 1/m * X_b.T.dot(sigmoid(X_b.dot(theta))-y)
    theta = theta - eta * gradient

X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
Xnewb = np.c_[np.ones((1000, 1)), X_new]
y_proba = sigmoid(Xnewb.dot(theta))

plt.scatter(X, y, marker="^")

plt.plot(X_new, y_proba, "b--", label="Iris virginica")
plt.subplot().set_xlabel("Sirina latice")
plt.subplot().set_ylabel("Verovatnoca")
plt.legend()
plt.show()
```

```python
print(y_proba)
```

```python
# Logistic regression using library

log_reg = LogisticRegression()
log_reg.fit(X, y)

X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)

plt.scatter(X, y, marker="^")
plt.plot(X_new, y_proba[:, 1], "g-", label="Iris virginica")
plt.plot(X_new, y_proba[:, 0], "b--", label="Nije Iris virginica")
plt.show()
```
